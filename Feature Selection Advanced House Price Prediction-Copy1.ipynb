### Feature Selection Advanced House Price Prediction

The main aim of the project is to predict the house price based on variours feature which will we discuss as we go ahead 

#### Dataset to download from the below link 

https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

```python
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline


## for feature selection 
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# To visualize the columns of the dataframe 
pd.pandas.set_option('Display.max_columns', None)

dataset=pd.read_csv('x_train.csv')
```

```python
dataset.head()
```

```python
# Capture the dependant features
y_train=dataset[['SalePrice']]
```

```python
# drop the dependant feature from the dataset
x_train=dataset.drop(['Id', 'SalePrice'],axis=1)
```

```python
### Apply feature se;ection 
# first, I specify the Lasso Regression model. amd I 
# Select a suitable alpha (equitable of poverty)
# The bigger the alpha the less features that will be selected

# Then i use the SelectFromModel fro sklearn, which 
# will select the features which coefficients are non-zero

# Impute missing values in x_train
imputer = SimpleImputer(strategy='mean')  # or 'median', etc.
x_train_imputed = imputer.fit_transform(x_train)

# Apply Lasso for feature selection
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train_imputed, y_train)


```

```python
feature_sel_model.get_support()
```

```python
# Let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat= x_train.columns[(feature_sel_model.get_support())]

# Let's print some stats 
print('total feature: {}'.format((x_train.shape[1])))
print('selected feature: {}'.format(len(selected_feat)))
print('features with selected shrank to zero: {}'. format(
   np.sum(feature_sel_model.estimator_.coef_ == 0)))
```
